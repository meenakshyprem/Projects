#1.which locations has rainfall of normal range 0.10

```python
#justification
df1=df[(df['Rainfall']>0.10) & (df['Rainfall']<0.25)]
df3=pd.pivot_table(
    data=df1,
    index='Location',
    values='Rainfall',
    aggfunc='mean'
).reset_index().sort_values(by='Rainfall',ascending=False).head()
print(df3)

```

```python
import seaborn as sns
sns.barplot(x='Location',y='Rainfall',data=df3,palette='husl')
import matplotlib.pyplot as plt
plt.title('Regions with average rainfall')
plt.xticks(rotation=45)
```

CONCLUSION: 
This figure displays those locations having average rainfall between 0.10 and 0.25 which is the avg rainfall value. It calculates the mean of the average rainfall values and shows the top 5 locations with highest average rainfall value.

#6 Correlation b/w Total and Average WindGust Speed in various locations and to check that location with higher windgust speed and increased rainfall

```python
agg_data=df.groupby('Location').agg(
    Total_Rainfall=('Rainfall','sum'),
    Avg_Windgustspeed=('WindGustSpeed','mean')
).reset_index()
```

```python
sns.barplot(x='Avg_Windgustspeed',y='Total_Rainfall',hue='Location',data=agg_data,palette='husl')
plt.title('Average Wind Gust Speed vs Total Rainfall by Location')
plt.xlabel('Average Wind Gust Speed')
plt.ylabel('Total Rainfall')
plt.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90,fontsize=7)
plt.show()
```

```python
'''Below Median: Locations where both wind gust speed and rainfall are below the central tendency.
Above Median: Locations where both wind gust speed and rainfall are above the central tendency, thus focusing on the more extreme cases of interest'''
```

```python
loc=agg_data[(agg_data['Avg_Windgustspeed']>agg_data['Avg_Windgustspeed'].median()) & (agg_data['Total_Rainfall']>agg_data['Total_Rainfall'].median())]
print(loc[['Location','Total_Rainfall','Avg_Windgustspeed']].sort_values(by=['Total_Rainfall','Avg_Windgustspeed'],ascending=False))#loc with largest wind gust speed and total rainfall
```

```python
sns.barplot(x='Avg_Windgustspeed',y='Total_Rainfall',hue='Location',data=loc,palette='husl')
plt.title('Average Wind Gust Speed vs Total Rainfall by Location')
plt.xlabel('Average Wind Gust Speed')
plt.ylabel('Total Rainfall')
plt.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45,fontsize=7)
plt.show()
```

CONCLUSION: 
Locations with above-median values in both total rainfall and average wind gust speed reveal that Gold Coast has higher rainfall, while Sydney Airport experiences stronger wind gusts.

CONCLUSION: 
The top 5 locations with the highest wind speeds show that Norah Head and Wollongong experience stronger winds at 9 am, while Gold Coast has higher wind speeds at 3 pm, indicating these areas are more susceptible to heavy rainfall.

#9 How average temperature and humidity at 9 AM and 3 PM vary across different locations and to draw relationship between them

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=True)
fig.suptitle("Top 5 locations with average Temperature at 9am and 3pm")
sns.barplot(ax=axes[0], x='Location', y='Temp9am',data=pd9t,palette='husl')
axes[0].set_title('Avg Temp at 9 am')
axes[0].tick_params(axis='x', rotation=45)
sns.barplot(ax=axes[1], x='Location', y='Temp3pm',data=pd9t,palette='husl')
axes[1].set_title('Avg Temp at 3 pm')
axes[1].tick_params(axis='x', rotation=45)
```

###conclusion:

The code calculates the top 5 locations with the highest average temperatures at 9 am and 3 pm, showing that Darwin has the highest averages (31.09°C at 3 pm and 27.16°C at 9 am), followed by Cairns, Townsville, Gold Coast, and Brisbane. The bar plots highlight these temperature trends across both times.

```python
#Calculate the average Humidity at 9 AM and 3 PM for each location.
pd9h=pd.pivot_table(
    data=df,
    index='Location',
    values=['Humidity9am','Humidity3pm'],
    aggfunc='mean'
).reset_index().sort_values(by=['Humidity9am','Humidity3pm'],ascending=[False,False]).head()
print(pd9h)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming pd9h contains the humidity data with columns 'Location', 'Humidity9am', 'Humidity3pm'

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Set the title for the overall figure
fig.suptitle("Top 5 Locations with Humidity at 9 AM and 3 PM")

# Scatter plot for Humidity at 9 AM
sns.scatterplot(ax=axes[0], x='Location', y='Humidity9am', data=pd9h, palette='husl',s=100)
axes[0].set_title('Humidity at 9 AM')
axes[0].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for readability

# Scatter plot for Humidity at 3 PM
sns.scatterplot(ax=axes[1], x='Location', y='Humidity3pm', data=pd9h, palette='husl',s=100)
axes[1].set_title('Humidity at 3 PM')
axes[1].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for readability

# Show the plots
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to accommodate the suptitle
plt.show()
```

###conclusion:

#11 To analyze the relationship between cloudiness at 9 AM and 3 PM, and how it relates to humidity and rainfall at various locations

```python
cloud_humidity_rainfall = df.groupby('Location')[['Cloud9am', 'Cloud3pm', 'Humidity9am', 'Humidity3pm', 'Rainfall']].mean().reset_index().sort_values(by=['Cloud9am','Cloud3pm'],ascending=[False,False]).head()
cloud_humidity_rainfall
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

cloud_humidity_rainfall['Location'] = cloud_humidity_rainfall['Location'].astype('category')

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Cloudiness vs Humidity at 9am
sns.barplot(ax=axes[0, 0], x='Cloud9am', y='Humidity9am', hue='Location', data=cloud_humidity_rainfall, palette='icefire')
axes[0, 0].set_title('Cloudiness vs Humidity at 9 AM')
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].legend(title='Location', loc='upper right', bbox_to_anchor=(1.05, 1), borderaxespad=0)

# Cloudiness vs Rainfall
sns.barplot(ax=axes[0, 1], x='Cloud9am', y='Rainfall', hue='Location', data=cloud_humidity_rainfall, palette='viridis')
axes[0, 1].set_title('Cloudiness vs Rainfall at 9 AM')
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].legend(title='Location', loc='upper right', bbox_to_anchor=(1.05, 1), borderaxespad=0)

# Cloudiness vs Humidity at 3pm
sns.barplot(ax=axes[1, 0], x='Cloud3pm', y='Humidity3pm', hue='Location', data=cloud_humidity_rainfall, palette='rocket')
axes[1, 0].set_title('Cloudiness vs Humidity at 3 PM')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].legend(title='Location', loc='upper right', bbox_to_anchor=(1.05, 1), borderaxespad=0)

# Cloudiness vs Rainfall at 3pm
sns.barplot(ax=axes[1, 1], x='Cloud3pm', y='Rainfall', hue='Location', data=cloud_humidity_rainfall, palette='mako')
axes[1, 1].set_title('Cloudiness vs Rainfall at 3 PM')
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].legend(title='Location', loc='upper right', bbox_to_anchor=(1.05, 1), borderaxespad=0)

plt.tight_layout()
plt.show()
```

CONCLUSION:
This analysis indicates the cloudiness affects humidity levels as well as rainfall. Regions like Gold Coast and Norah Head show a strong correlation, while regions like Penrith and Salmon Gums exhibit more variable weather patterns. 
 Cloudiness, humidity, and rainfall show a complex relationship where higher cloud cover often corresponds with increased humidity and rainfall, but the strength of these correlations can vary across different times of the day and locations.

#16 To identify those locations whose minimum and maximum temperature is greater than their average

```python

df_mintemp = df['MinTemp'].mean()
df_maxtemp = df['MaxTemp'].mean()
df_filtered = df[(df['MinTemp'] > df_mintemp) & (df['MaxTemp'] > df_maxtemp)]
df_filtered_grouped = df_filtered.sort_values(by=['MinTemp', 'MaxTemp'], ascending=False).drop_duplicates(subset='Location').head()
print(df_filtered_grouped[['Location', 'MinTemp', 'MaxTemp']])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.barplot(data=df_filtered_grouped, x='Location', y='MinTemp', palette='Set1', ax=axes[0])
sns.barplot(data=df_filtered_grouped, x='Location', y='MaxTemp', palette='Set2',ax=axes[1])
axes[0].set_title('Min Temperature Above Average')
axes[0].set_xlabel('Location')
axes[0].set_ylabel('Min Temperature')
axes[1].set_title('Max Temperature Above Average')
axes[1].set_xlabel('Location')
axes[1].set_ylabel('Max Temperature')
plt.tight_layout()
plt.show()
```

CONCLUSION: 
The filtered locations indicate areas consistently experiencing higher-than-average temperatures, suggesting a hotter climate that may lead to heat-related challenges and affect local ecosystems and human activities.
